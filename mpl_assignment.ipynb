```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
a = list(np.random.randint(1000, size=50))
b = list(np.random.randint(1000, size=50))

a_even = [a[i] for i in range(len(a)) if (a[i]+b[i]) %2 == 0]
b_even = [b[i] for i in range(len(a)) if (a[i]+b[i]) %2 == 0]

a_odd = [a[i] for i in range(len(a)) if (a[i]+b[i]) %2 != 0]
b_odd = [b[i] for i in range(len(a)) if (a[i]+b[i]) %2 != 0]
```

```python
plt.scatter(a_even,b_even,color='r',label='even')
plt.scatter(a_odd,b_odd,color='b',label='odd')
plt.legend()
```

```python
def func1(x):
    return 3*x+.5

def func2(x):
    import math as mt
    return 5*mt.sqrt(x)

x1_values = np.arange(0,6)
y1_values = [func1(i) for i in x1_values]
y1_values

y2_values = [func2(i) for i in x1_values]
```

```python
fig,ax = plt.subplots(1,1,figsize=(5,5))
ax.plot(x1_values,y1_values, color = 'r', linestyle = '-', marker = '*', label = "func1")
ax.plot(x1_values,y2_values, color = 'b', linestyle = '--', marker = '^', label = "func2")
ax.legend()
```

```python
fig,ax = plt.subplots(1,1,figsize=(5,5))
ax.plot(x1_values,y1_values, color = 'r', linestyle = '-', marker = '*', label = "func1")
ax.plot(x1_values,y2_values, color = 'b', linestyle = '--', marker = '^', label = "func2")
ax.legend()
ax.set_xscale("log")
```

```python
barheights = [5,3,1]
barlabels = ['oranges', 'grapes', 'hockey pucks']
fig, ax = plt.subplots()
ax.bar(np.arange(len(barheights)), barheights)
x_pos = np.arange(len(barheights))
ax.set_xticks(x_pos)
ax.set_xticklabels(barlabels, rotation=45)
```

```python
fig,ax = plt.subplots(2,2,figsize=(8,5))

ax[0,0].scatter(a_even,b_even,color='r',label='even')
ax[0,0].scatter(a_odd,b_odd,color='b',label='odd')
ax[0,0].legend(loc='upper right')


ax[0,1].plot(x1_values,y1_values, color = 'r', linestyle = '-', marker = '*', label = "func1")
ax[0,1].plot(x1_values,y2_values, color = 'b', linestyle = '--', marker = '^', label = "func2")
ax[0,1].legend()

ax[1,0].plot(x1_values,y1_values, color = 'r', linestyle = '-', marker = '*', label = "func1")
ax[1,0].plot(x1_values,y2_values, color = 'b', linestyle = '--', marker = '^', label = "func2")
ax[1,0].legend()
ax[1,0].set_xscale("log")

ax[1,1].bar(np.arange(len(barheights)), barheights)

plt.savefig("mpl_assignment_4plots.png")
```
